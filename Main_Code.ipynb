```python
problem = get_problem("zdt1")

# Define the range of input values for evaluation
n = 100  # Number of points to generate
x = np.linspace(0, 1, n)
X = np.meshgrid(x, x)

# Flatten the input values and evaluate the problem
F = np.empty((n, problem.n_var))
F[:, 0] = x
    
print(len(F))

points = problem.evaluate(F)

plot_pts = np.empty((100,2))

for p in points:
    plot_pts.append(p)
    
plt.plot(plot_pts[:,0], plot_pts[:,1])
```
